Was reminded of the importance of checking data for outliers by this article, and modeled my filtering code below on this article:
https://www.analyticsvidhya.com/blog/2021/08/holt-winters-method-for-time-series-analysis/

```python
sns.boxplot(x = myrbetriq['murb_dollars']);
```

```python
myrbetriq['z_score'] = [ ((i - myrbetriq['murb_dollars'].mean())/myrbetriq['murb_dollars'].std()) for i in myrbetriq['murb_dollars']]
```

```python
myrbetriq.shape
```

```python
myrbetriqb = myrbetriq[(np.abs(myrbetriq['z_score']) < 3)].copy()
```

```python
myrbetriqb.shape
```

```python
myrbetriq_monthly = myrbetriqb.resample('m').sum()
```

```python
myrbetriq_monthly.head()
```

```python
myrbetriq_monthly.sort_index()
```

Adding 1 to the weekly sales for each observation so that Holt_winters will work.

Baseline - Simple Mean (Naive Model)

```python
df_preds = test.copy(deep=True)
df_preds
```

```python
df_preds['simple_mean'] = train['murb_dollars'].mean()
df_preds
```

```python
plt.figure(figsize=(12,8))

plt.plot(train.index, train['murb_dollars'], label='Train')
plt.plot(test.index, test['murb_dollars'], label='Test')
plt.plot(df_preds.index, df_preds['simple_mean'], label="Baseline (Mean)")

plt.legend(loc='best')
plt.title("Baseline (Mean) Forecast for Monthly Myrbetriq Payments");
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
print("MAE for Simple Mean Baseline Model:", mean_absolute_error(df_preds['murb_dollars'], df_preds['simple_mean']))
print("RMSE for Simple Mean Baseline Model:", mean_squared_error(df_preds['murb_dollars'], df_preds['simple_mean'], squared=False))
```

Make a Seasonal Baseline Model with shift of 7

```python
df_preds['seasonal_7'] = train['murb_dollars'].shift(7, freq='M') 
df_preds.head(2)
```

```python
plt.figure(figsize=(12,8))

plt.plot(train.index, train['murb_dollars'], label='Train')
plt.plot(test.index, test['murb_dollars'], label='Test')
plt.plot(df_preds.index, df_preds['seasonal_7'], label='Seasonal 7')

plt.legend(loc='best')
plt.title("Seasonal 7 Forecast for Monthly Myrbetriq Payments");
```

```python
df_preds['seasonal_7'].fillna(0, inplace=True)
```

```python
print("MAE for Seasonal Baseline 7 Shift Model:", mean_absolute_error(df_preds['murb_dollars'], df_preds['seasonal_7']))
print("RMSE for Seasonal Baseline 7 Shift Model:", mean_squared_error(df_preds['murb_dollars'], df_preds['seasonal_7'], squared=False))
```

Simple Exponential Smoothing Model

```python
from statsmodels.tsa.api import SimpleExpSmoothing, ExponentialSmoothing
```

```python
ses = SimpleExpSmoothing(train['murb_dollars']).fit(smoothing_level=.5)
```

```python
ses.summary()
```

```python
ses_forecast = ses.forecast(len(test))
```

```python
df_preds['ses'] = ses_forecast
```

```python
plt.figure(figsize=(12,8))

plt.plot(train.index, train['murb_dollars'], label='Train')
plt.plot(test.index, test['murb_dollars'], label='Test')
plt.plot(df_preds.index, df_preds['ses'], label='Simple Exponential Smoothing')

plt.legend(loc='best')
plt.title("Simple Exponential Smoothing for Myrbetriq Monthly Payments");
```

```python
print("MAE for Simple Exponential Smoothing Model:", mean_absolute_error(df_preds['murb_dollars'], df_preds['ses']))
print("RMSE for Simple Exponential Smoothing Model:", mean_squared_error(df_preds['murb_dollars'], df_preds['ses'], squared=False))
```

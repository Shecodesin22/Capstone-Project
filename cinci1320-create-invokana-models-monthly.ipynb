Was reminded of the importance of checking data for outliers by this article, and modeled my filtering code below on this article:
https://www.analyticsvidhya.com/blog/2021/08/holt-winters-method-for-time-series-analysis/

```python
sns.boxplot(x = invokana['invok_dollars']);
```

```python
invokana['z_score'] = [ ((i - invokana['invok_dollars'].mean())/invokana['invok_dollars'].std()) for i in invokana['invok_dollars']]
```

```python
invokana.shape
```

Add chart to see data before dropping outliers

```python
plt.figure(figsize=(12,8))

plt.plot(invokana.index, invokana['invok_dollars'], label='Before Dropping Outliers')


plt.legend(loc='best')
plt.title("Monthly Payments for Invokana Drug - Before Outliers Dropped");
```

```python
invokanab = invokana[(np.abs(invokana['z_score']) < 3)].copy()
```

```python
invokanab.shape
```

Add chart to see data after dropping outliers

```python
plt.figure(figsize=(12,8))

plt.plot(invokanab.index, invokanab['invok_dollars'], label='After Dropping Outliers')


plt.legend(loc='best')
plt.title("Monthly Payments for Invokana Drug - After Outliers Dropped");
```

Baseline - Simple Mean (Naive Model)

```python
df_preds = test.copy(deep=True)
df_preds
```

```python
df_preds['simple_mean'] = train['invok_dollars'].mean()
df_preds
```

```python
plt.figure(figsize=(12,8))

plt.plot(train.index, train['invok_dollars'], label='Train')
plt.plot(test.index, test['invok_dollars'], label='Test')
plt.plot(df_preds.index, df_preds['simple_mean'], label="Baseline (Mean)")

plt.legend(loc='best')
plt.title("Baseline (Mean) Forecast for Monthly Invokana Payments");
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
print("MAE for Simple Mean Baseline Model:", mean_absolute_error(df_preds['invok_dollars'], df_preds['simple_mean']))
print("RMSE for Simple Mean Baseline Model:", mean_squared_error(df_preds['invok_dollars'], df_preds['simple_mean'], squared=False))
```

Make a Seasonal Baseline Model with shift of 4

```python
df_preds['seasonal_4'] = train['invok_dollars'].shift(4, freq='M') 
df_preds.head(2)
```

```python
plt.figure(figsize=(12,8))

plt.plot(train.index, train['invok_dollars'], label='Train')
plt.plot(test.index, test['invok_dollars'], label='Test')
plt.plot(df_preds.index, df_preds['seasonal_4'], label='Seasonal 4')

plt.legend(loc='best')
plt.title("Seasonal 4 Forecast for Monthly Invokana Payments");
```

```python
df_preds['seasonal_4'].fillna(0, inplace=True)
```

```python
print("MAE for Seasonal Baseline 4 Shift Model:", mean_absolute_error(df_preds['invok_dollars'], df_preds['seasonal_4']))
print("RMSE for Seasonal Baseline 4 Shift Model:", mean_squared_error(df_preds['invok_dollars'], df_preds['seasonal_4'], squared=False))
```

Simple Exponential Smoothing Model

```python
from statsmodels.tsa.api import SimpleExpSmoothing, ExponentialSmoothing
```

```python
ses = SimpleExpSmoothing(train['invok_dollars']).fit(smoothing_level=.5)
```

```python
ses.summary()
```

```python
ses_forecast = ses.forecast(len(test))
```

```python
df_preds['ses'] = ses_forecast
```

```python
plt.figure(figsize=(12,8))

plt.plot(train.index, train['invok_dollars'], label='Train')
plt.plot(test.index, test['invok_dollars'], label='Test')
plt.plot(df_preds.index, df_preds['ses'], label='Simple Exponential Smoothing')

plt.legend(loc='best')
plt.title("Simple Exponential Smoothing for Invokana Monthly Payments");
```

```python
print("MAE for Simple Exponential Smoothing Model:", mean_absolute_error(df_preds['invok_dollars'], df_preds['ses']))
print("RMSE for Simple Exponential Smoothing Model:", mean_squared_error(df_preds['invok_dollars'], df_preds['ses'], squared=False))
```
